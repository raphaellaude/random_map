```python
def get_random_location(shape):
    x = np.random.randint(0, shape[0])
    y = np.random.randint(0, shape[1])
    
    return x, y
```

```python
get_random_location((10, 10))
```

```python
# Set-up
shape = (25, 100)

food = np.zeros(shape)

for x in range(20):
    fx, fy = get_random_location(shape)
    food[fx, fy] = 5

petri = slime.PetriDish(shape, food, pheromone_evaporation=0.05)

options = {
    'shape':shape,
    'forward_bias':0.25,
    'p_strength':1,
    'view_distance':4
}

# Run
agents = [slime.Agent(home=np.array(get_random_location(shape)), **options) for x in range(10)]
petri.grow_slime(agents)
```

```python
slime.draw(petri.slime);
```
